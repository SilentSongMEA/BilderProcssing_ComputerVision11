```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
model=cv2.dnn.readNetFromTensorflow(r'C:\Users\Student\Desktop\pic11\frozen_inference_graph.pb',r'C:\Users\Student\Desktop\pic11\ssd_mobilenet_v2_coco_2018_03_29.pbtxt')
with open (r'C:\Users\Student\Desktop\pic11\object_detection_classes_coco.txt','r') as f:
    class_names=f.read().split('\n')
COLORS=np.random.uniform(0,255,size=(len(class_names),3))
image=cv2.imread(r"C:\Users\Student\Desktop\pic11\00.jpg")
image_height,image_width,_=image.shape

model.setInput(cv2.dnn.blobFromImage(image,size=(300,300),swapRB=True,crop=False))
output=model.forward()

for detection in output[0,0,:,:]:
    score=float(detection[2])
    if score > 0.3:
        class_id=detection[1]
        class_name=class_names[int(class_id)-1]
        color=COLORS[int(class_id)]

        left=detection[3] * image_width
        top=detection[4] * image_height
        right=detection[5] * image_width
        bottom=detection[6] * image_height

        cv2.rectangle(image,(int(left), int(top)), (int(right), int(bottom)), color,  thickness=2)
        cv2.putText(image,class_name, (int(left), int(top - 5)), cv2.FONT_HERSHEY_SIMPLEX, 0.4, color, 2)
        print(class_name, round(score*100,2),"%")

plt.imshow(image[...,::-1])
```
