```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
synset_file=r"C:\Users\Student\Desktop\pic11\synset_words.txt"
prototext_file=r"C:\Users\Student\Desktop\pic11\bvlc_googlenet.prototxt"
caffemodel_file=r"C:\Users\Student\Desktop\pic11\bvlc_googlenet.caffemodel"
image_file=r"C:\Users\Student\Desktop\pic11\elephant.jpg"
```

```python
with open(synset_file,"r") as f:
    rows=f.read().strip().split("\n")
classes= [r[r.find("")+1:].split(",")[0] for r in rows]
```

```python
image=cv2.imread(image_file)
```

```python
if image is None:
    raise FileNotFoundError(f"Image not found: {image_file}")
image_rgb=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
blob=cv2.dnn.blobFromImage(image_rgb,1,(224,224),(104,117,123))
```

```python
net=cv2.dnn.readNetFromCaffe(prototext_file, caffemodel_file)
net.setInput(blob)
preds=net.forward()
idx=np.argmax(preds)
confidence=preds[0][idx]
label=classes[idx]
print("Predicted: {} with confidence: {:.2f}%".format(label, confidence * 100))
plt.imshow(image_rgb)
plt.title("{} {:.2f}%".format(label, confidence * 100))
plt.axis('off')
plt.show()
```
